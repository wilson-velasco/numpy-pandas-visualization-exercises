```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint

# 1. Use matplotlib to plot the following equation:

# y = x**2 âˆ’ x + 2

# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.

x = [randint(0,50) + x for x in range(20)]
y = [((t ** 2) - t + 2) for t in x]
```

```python
plt.scatter(x,y)
plt.xlim(0,100)
plt.xlabel('x')
plt.ylabel('y')
plt.annotate('Origin', xy=(0, 0), xytext=(20, 1000),
             arrowprops={'facecolor': 'red'});
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y = x ** -2
# y = x ** 3
# y = 2 ** x
# y = 1 / (x+1)

# You can use functions from the math module to help implement some of the equations above.

from random import uniform

x1 = [uniform(1,5) for x in range(20)]
y1 = [x ** .5 for x in x1]

x2 = [uniform(-2,2) for x in range(20)]
y2 = [x ** 3 for x in x2]

x3 = [uniform(-3,3) for x in range(20)]
y3 = [2 ** x for x in x3]

x4 = [uniform(-5,5) for x in range(20)]
y4 = [1 / (x + 1) for x in x4]
```

```python
plt.scatter(x1,y1, label=1)
plt.scatter(x2,y2, label=2)
plt.scatter(x3,y3, label=3)
plt.scatter(x4,y4, label=4)

plt.legend();
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.

plt.subplot(221)
plt.scatter(x1,y1, label=1)
plt.legend()

plt.subplot(222)
plt.scatter(x2,y2, label=2)
plt.legend()

plt.subplot(223)
plt.scatter(x3,y3, label=3)
plt.legend()

plt.subplot(224)
plt.scatter(x4,y4, label=4)
plt.legend()
```

```python
# 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize=(10,10))

plt.subplot(221)
plt.scatter(x1,y1, label=1, c='green')
plt.title(r'$y = \sqrt{x}$')
plt.legend()

plt.subplot(222)
plt.scatter(x2,y2, label=2, c='blue')
plt.title(r'$y = x^{3}$')
plt.legend()

plt.subplot(223)
plt.scatter(x3,y3, label=3, c='magenta')
plt.title(r'$y = 2^{x}$')
plt.legend()

plt.subplot(224)
plt.scatter(x4,y4, label=4, c='orange')
plt.title(r'$y = \frac{1}{x+1}$')
plt.legend();
```

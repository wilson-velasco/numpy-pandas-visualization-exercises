```python
import pandas as pd
import numpy as np
from random import uniform
from random import randint
from math import log
from math import factorial

x1 = [uniform(1,5) for x in range(20)]
y1 = [0*x + 1 for x in x1]

x2 = [uniform(0,5) for x in range(20)]
y2 = [log(x) for x in x2]

x3 = [uniform(0,5) for x in range(20)]
y3 = [x * log(x) for x in x3]

x4 = [uniform(-5,5) for x in range(20)]
y4 = [x ** 2 for x in x4]

x5 = [randint(0,4) for x in range(20)]
y5 = [factorial(x) for x in x5]

x6 = [uniform(-2,2) for x in range(20)]
y6 = [x ** x  for x in x6]
```

```python
plt.figure(figsize=(15,15))

plt.subplot(321)
plt.scatter(x1,y1, label=1, c='green')
plt.title(r'$y = 0n + 1$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.subplot(322)
plt.scatter(x2,y2, label=2, c='blue')
plt.title(r'$y = \log(n)$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.subplot(323)
plt.scatter(x3,y3, label=3, c='magenta')
plt.title(r'$y = n \times \log(n)$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.subplot(324)
plt.scatter(x4,y4, label=4, c='orange')
plt.title(r'$y = x^{2}$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.subplot(325)
plt.scatter(x5,y5, label=5, c='skyblue')
plt.title(r'$y = n!$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.subplot(326)
plt.scatter(x6,y6, label=6, c='red')
plt.title(r'$y = x^{x}$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend();
```
